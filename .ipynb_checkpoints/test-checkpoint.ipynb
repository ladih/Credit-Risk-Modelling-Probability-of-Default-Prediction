# Credit Risk Modelling: Predicting Probability of Default Using Logistic Regression


#### Goal: Given a borrower’s loan data, predict the probability they will default on the loan.

This project uses the Lending Club loan dataset (https://www.kaggle.com/datasets/adarshsng/lending-club-loan-data-csv) to build a probability of default (PD) estimator and evaluate its performance.

## Step 1: Initial preparation

#### Libraries
Import essential libraries for data manipulation, preprocessing, logistic regression modeling, performance evaluation, and visualization.  

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve, roc_auc_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
```

#### Load pre-processed dataset
Use a smaller working sample (loan_sample.csv) that preserves the original proportion of default vs. non-default cases.

```python
df = pd.read_csv('loan_sample.csv', low_memory=False)
print(f"Data consists of {df.shape[0]} rows and {df.shape[1]} columns.")
```

#### Filter target variable

Loans with status "Charged Off" or "Default" are labeled as **default** and "Fully Paid" as **non-default**. Other statuses (e.g., Current, Late (16–120 days)) were excluded.

```python
target_statuses = ['Fully Paid', 'Charged Off', 'Default']
df_filtered = df[df['loan_status'].isin(target_statuses)].copy()
print(f"Shape after filtering target statuses: {df_filtered.shape[0]} rows, {df_filtered.shape[1]} columns.")
```

#### Create binary target

Encode defaulted loans as 1 and non-defaulted loans as 0.

```python
df_filtered['default'] = df_filtered['loan_status'].isin(['Charged Off', 'Default']).astype(int)
```

#### Class distribution

```python
counts = df_filtered['default'].value_counts()
percent = counts / counts.sum() * 100

for cls in counts.index:
    print(f"Class {cls}: {counts[cls]} ({percent[cls]:.2f}%)")    
```

## Step 2: Variable selection

Only variables definitely available at loan origination were selected. Variables with uncertain availability were excluded, resulting in the following set (descriptions from LCDataDictionary.xlsx):

- **annual_inc**         - The self-reported annual income provided by the borrower during registration.

- **application_type**   - Indicates whether the loan is an individual application or a joint application with two co-borrowers.

- **dti**                - A ratio calculated using the borrower’s total monthly debt payments on the total debt obligations, excluding mortgage and the requested LC loan, divided by the borrower’s self-reported monthly income.

- **home_ownership**     - The home ownership status provided by the borrower during registration or obtained from the credit report. Our values are: RENT, OWN, MORTGAGE, OTHER.

- **purpose**            - A category provided by the borrower for the loan request.

Example of variables with uncertain time measurments:

- **delinq_2yrs**   - The number of 30+ days past-due incidences of delinquency in the borrower's credit file for the past 2 years.  
    *(timing relative to loan origin is unclear)*

- **emp_length**    - Employment length in years. Possible values are between 0 and 10 where 0 means less than one year and 10 means ten or more years.  
  *(may have been updated after loan origination)*

Extra variables:

- **loan_status** - Current status of the loan  
    *Note:* Target variable.
- **issue_d** - The month which the loan was funded  
    *Note:* Used only for time-based splitting.
    

```python
selected_vars = [
    'annual_inc',
    'application_type',
    'dti',
    'home_ownership',
    'purpose',
    'default',  # target
    'issue_d',  # for time-based splitting
]
df_model = df_filtered[selected_vars].copy()
print(f"Shape after variable selection: {df_model.shape[0]} rows, {df_model.shape[1]} columns.")
```

#### Check numeric variables

```python
df_model.describe()
```

**Note:** Maximum of dti is 999. While unusually high, these cases correspond to borrowers with very low reported income (mean = 430 compared to overall mean = 75,989, so the values are retained in the dataset.


## Step 2: Data Preprocessing

- Categorical variables are one-hot encoded. 
- Missing values are imputed using the median for each feature.
- Train/test split is time-based: the first 80% of loans were used for training, the last 20% for testing.
- Feature scaling: Features are standardized to have mean 0 and standard deviation 1.

#### Identify variables with missing values

```python
missing_counts = df_model.isna().sum().sort_values(ascending=False)
total_rows = len(df_model)

missing_vars = missing_counts[missing_counts > 0]

if len(missing_vars) > 0:
    for col, count in missing_vars.items():
        percentage = (count / total_rows) * 100
        print(f"{col}: {count} ({percentage:.2f}%)")
else:
    print("No missing values found")
```

#### Replace missing values with median for each numeric column

```python
numeric_cols = df_model.select_dtypes(include=['number']).columns
df_model[numeric_cols] = df_model[numeric_cols].fillna(df_model[numeric_cols].median())
```

#### One-hot encode categorical variables

```python
categorical_cols = ['purpose', 'home_ownership', 'application_type']

df_model = pd.get_dummies(df_model, columns=categorical_cols, drop_first=True)
```

```python
print(f"Shape after one-hot encoding: {df_model.shape[0]} rows, {df_model.shape[1]} columns.")
```

#### Time-based split of train and test sets

First 80% of the sorted data is used for training, the last 20% for testing.

```python
df_sorted = df_model.sort_values('issue_d')

split_point = int(0.8 * len(df_sorted))
# Check split point
train_start = df_sorted.iloc[:split_point]['issue_d'].iloc[0]
train_end = df_sorted.iloc[:split_point]['issue_d'].iloc[-1]
test_start = df_sorted.iloc[split_point:]['issue_d'].iloc[0]
test_end = df_sorted.iloc[split_point:]['issue_d'].iloc[-1]
print(f"Training period: {train_start} to {train_end}")
print(f"Test period: {test_start} to {test_end}")

X_train = df_sorted.iloc[:split_point].drop(['default', 'issue_d'], axis=1)
y_train = df_sorted.iloc[:split_point]['default']
X_test = df_sorted.iloc[split_point:].drop(['default', 'issue_d'], axis=1)
y_test = df_sorted.iloc[split_point:]['default']

print(f"\nTraining set: {X_train.shape[0]} loans")
print(f"Test set: {X_test.shape[0]} loans")
print(f"Training default rate: {y_train.mean():.3f}")
print(f"Test default rate: {y_test.mean():.3f}")
```

#### Feature scaling

Scale features so each feature has mean 0 and standard deviation 1. This makes the coefficients in logistic regression easier to compare.

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

## Step 3: Model Training

A logistic regression model is trained. The assumption is that the probability of default is given by

p(x1, ..., xm) = 1 / (1 + exp(-(b0 + b1 * x1 + ... + bm * xm))),

where x1, ..., xm are the input features, and b0, ..., bm are the coefficients learned during training.

```python
# Initialize model
model = LogisticRegression()

# Train the model on scaled training data (uses L2 regularization by default)
model.fit(X_train_scaled, y_train);
```

#### Make predictions

```python
# Predict probabilities of default
y_pred_proba = model.predict_proba(X_test_scaled)[:, 1]

# Convert probabilities to binary predictions
threshold = 0.5
y_pred = (y_pred_proba >= threshold).astype(int)
```

## Step 4: Result Metrics

### Threshold-dependent metrics (threshold=0.5)

The classification report shows precision, recall, and f1-score for non-defaulters (0) and defaulters (1).

```python
print(classification_report(y_test, y_pred))
```

**Comments:**  
From these metrics, we see that the model correctly classifies 100% of non-defaulters and 0% of defaulters (recall). One reason for this is the combination of the **imbalanced class proportions** (20% default) and the **balanced threshold** (50%).

One way to achieve a more balanced recall is to lower the classification threshold. For example, a threshold of 0.2 could be reasonable, as it reflects the underlying class imbalance. (With a threshold of 0.2 the recalls are indeed more balanced: 0.56 and 0.58.)


### Threshold independent metrics


#### AUC and Gini Coefficient
The AUC (Area Under the ROC Curve) represents the probability that the model assigns a higher score (i.e., higher predicted probability) to a randomly chosen defaulter than to a randomly chosen non-defaulter. A random classifier would do this 50% of the time and hence have an AUC of 0.5. Values closer to 1 indicate better discrimination between the classes.

The Gini coefficient is defined as 2 × AUC − 1. It provides no additional information beyond the AUC but is commonly reported in credit risk practice.

```python
auc = roc_auc_score(y_test, y_pred_proba)
print(f"\nAUC: {auc:.3f}")

gini = 2 * auc - 1
print(f"Gini coefficient: {gini:.3f}")
```

The AUC of 0.605 indicates that the model performs better than random guessing, although in credit risk applications, higher values are generally preferred (e.g., >0.7). Using additional relevant features could improve the result.

#### KS statistic (Kolmogorov-Smirnov)

The KS statistic represents the maximum separation between the true positive rate (TPR) and false positive rate (FPR) across all thresholds. Values closer to 1 indicate stronger discrimination between the classes, with higher values suggesting the model is better at distinguishing defaulters from non-defaulters.

```python
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
ks = max(tpr - fpr)
print(f'KS: {ks:.3f}')
```

The KS statistic of 0.156 suggests that the model has weak discrimination between defaulters and non-defaulters, below typical industry thresholds for moderate separation.

### Stability/Monitoring metrics

#### PSI (Population Stability Index)

The Population Stability Index (PSI) measures how the distribution of predicted probabilities (or of individual features) changes between the training set and test set (or future/out-of-time samples). Values close to 0 indicate stable distributions, while larger values suggest population drift and potential model degradation. The PSI depends on how the data is partitioned into bins, and different binning strategies may slightly affect the value.

```python
# Predicted probabilities for the training set
y_pred_proba_train = model.predict_proba(X_train_scaled)[:, 1]

n_bins = 10
bin_edges = np.linspace(0, 1, n_bins + 1)

df_PSI_test = pd.DataFrame({'y_pred_proba': y_pred_proba})

df_PSI_test['bin'] = pd.cut(df_PSI_test['y_pred_proba'], bins=bin_edges, include_lowest=True)

df_test_bins = (
    df_PSI_test.groupby('bin')
    .agg(count=('y_pred_proba', 'count'))
    .reset_index()
)

df_test_bins['fraction_test'] = df_test_bins['count'] / df_test_bins['count'].sum()

df_PSI_train = pd.DataFrame({'y_pred_proba_train': y_pred_proba_train})

df_PSI_train['bin'] = pd.cut(df_PSI_train['y_pred_proba_train'], bins=bin_edges, include_lowest=True)

df_train_bins = (
    df_PSI_train.groupby('bin')
    .agg(count=('y_pred_proba_train', 'count'))
    .reset_index()
)

df_train_bins['fraction_train'] = df_train_bins['count'] / df_train_bins['count'].sum()

# psi = sum( [train_pred % - test_pred %] * ln (train_pred % / future_pred %) )
def calculate_psi(train_frac, test_frac):
    """
    train_frac: array of fractions per bin in the baseline population
    test_frac: array of fractions per bin in the new population
    """
    epsilon = 1e-8  # to avoid division by zero
    psi_values = (train_frac - test_frac) * np.log((train_frac + epsilon) / (test_frac + epsilon))
    return np.sum(psi_values)

psi = calculate_psi(df_train_bins['fraction_train'], df_test_bins['fraction_test'])
print(f'PSI ({n_bins} bins): {psi:.3f}')
```

The small PSI value indicates that the distributions of predicted probabilities for the train and test sets are very similar. This is confirmed by the plot below. From the plot we can also see that the model predicts most probabilities between 0 and 0.5, reflecting the class imbalance (80% of samples are class 0 (non-default)).

```python
# "PSI-plot"
plt.figure(figsize=(8,6))
x_mid = df_test_bins['bin'].apply(lambda x: x.mid).to_numpy()
plt.plot(x_mid, df_train_bins['fraction_train'].to_numpy(), label='Train', marker='o')
plt.plot(x_mid, df_test_bins['fraction_test'].to_numpy(), label='Test', marker='o')

plt.xticks(rotation=45)
plt.xticks(np.arange(0, 1.01, 0.1))  # ticks at 0.0, 0.1, 0.2, ..., 1.0
plt.xlabel("Probability bin")
plt.ylabel("Fraction of samples")
plt.title("Distribution of predicted probabilities (Train vs Test)")
plt.legend()
plt.tight_layout()
plt.show()
```

### Calibration & Probability Assessment


```python
df_calibration = pd.DataFrame({
    'y_test': y_test,
    'y_pred_proba': y_pred_proba
})

n_bins_cali = 10
bin_edges = np.linspace(0, 1, n_bins_cali + 1)

df_calibration["bin"] = pd.cut(df_calibration['y_pred_proba'], bins=bin_edges, include_lowest=True)
calibration_table = df_calibration.groupby('bin').agg(
    avg_pred=('y_pred_proba', 'mean'),          # Average predicted PD in the bin
    obs_default_rate=('y_test', 'mean'),        # Fraction of defaults in the bin
).reset_index() # make the bins a column instead of index, and use standard index

plt.figure(figsize=(8,6))
plt.plot([0,1], [0,1], 'k--', label='Perfect calibration')
plt.plot(calibration_table['avg_pred'].to_numpy(), calibration_table['obs_default_rate'].to_numpy(), marker='o', label='Model')
plt.xlabel('Average Predicted PD')
plt.ylabel('Observed Default Rate')
plt.title('Calibration Plot')
plt.legend()
plt.show()
```

Calibration plot: Shows how well predicted probabilities correspond to the actual fraction of defaulters. For example, if the average predicted probability in the bin [0, 0.1] is 0.06, the fraction of actual defaulters in that bin should also be approximately 0.06.

The model is well-calibrated for lower predicted probabilities, where most samples lie. At higher probabilities, there are few observations, and due to class imbalance (more non-defaulters), a few high-probability non-defaults can distort the observed fraction of defaulters. This explains the deviation from the diagonal in the high-probability region.

### Feature importance

A positive coefficient means the probability of default increases if the feature increases in value. A negative coefficient means the probability of default decreases if the feature increases. The table below shows the top 5 most influential features.

```python
feature_importance = pd.DataFrame({
    'feature': X_train.columns,
    'coefficient': model.coef_[0].round(3)
}).sort_values('coefficient', key=abs, ascending=False)
top_n_features = 5
feature_importance.head(top_n_features)

top_features = feature_importance.head(top_n_features)
top_features.style.hide(axis='index')
```

### Step 5. Visualizations

- **ROC curve including KS point:** Plots false positive rates (FPR) vs true positive rates (TPR) across thresholds. The KS statistic indicates the maximum separation between TPR and FPR.
- **Calibration plot:** Compares the actual fraction of defaulters with the average predicted probabilities for different probability ranges, assessing probability accuracy.
- **Predicted probability distributions (train vs test):** Compares the distribution of predicted probabilities between training and testing sets; useful for detecting potential population drift.

```python
plt.figure(figsize=(8,6))
# plot ROC curve
plt.plot(fpr, tpr, color='blue', lw=2, label=f'ROC curve (AUC = {auc:.3f})')

# plot KS point
ks_index = (tpr - fpr).argmax()
plt.scatter(fpr[ks_index], tpr[ks_index], color='red', s=30, label='KS point', zorder=5)

plt.plot([0, 1], [0, 1], color='gray', lw=1, linestyle='--')  # random classifier line
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
```

Standard ROC plot. According to the plot, the AUC value of 0.605 seems reasonable.
